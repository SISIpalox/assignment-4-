```python
import random

import matplotlib.pyplot as plt

steps = 25 

def simulate_touch_frequency(steps=25, trials=1000, tosses=24):
    """
    Simulate the coin flip game for a given number of trials and record the number of times each step is touched.
    Each game consists of a fixed number of coin tosses.
    Returns a dictionary with steps as keys and the count of times each step is touched as values.
    """
    touch_frequency = {step: 0 for step in range(1, steps + 1)}  # Initialize touch frequency dictionary

    for _ in range(trials):
        touched_steps = set()  # To record the steps that were touched in this trial
        position = 0  # Start at the beginning of the path
        for _ in range(tosses):  # Perform a fixed number of coin tosses
            flip = random.choice([1, 2])  # Simulate the coin flip
            position += flip  # Take the steps based on the coin flip
            # If the position is within our step range, record it
            if position <= steps:
                touched_steps.add(position)
        
        # Update the touch frequency for each touched step
        for step in touched_steps:
            touch_frequency[step] += 1

    return touch_frequency

# Simulate the game 1000 times with 24 tosses each and get the touch frequency for each step
touch_frequency = simulate_touch_frequency()

# Plot the touch frequency of each step
plt.figure(figsize=(12, 6))
plt.bar(touch_frequency.keys(), touch_frequency.values(), color='lightcoral')
plt.xlabel('Step Number')
plt.ylabel('Touch Frequency')
plt.title('Touch Frequency of Each Step in 1000 Games (24 tosses each)')
plt.xticks(range(1, steps + 1))
plt.grid(axis='y')
plt.tight_layout()

# Show the plot
plt.show()
```
